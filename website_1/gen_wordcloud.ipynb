```python
def strip_text(df):
    for i in range(len(df)):
        df['title'][i] = df['title'][i].replace('| Technology | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Society | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Cities | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Science | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Politics | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Global development | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Law | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Culture | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Sport | The Guardian', '')
        df['title'][i] = df['title'][i].replace('| Environment | The Guardian', '')
        df['title'][i].strip()
```

```python
strip_text(tech)
strip_text(politics)
strip_text(sport)
strip_text(gb)
strip_text(society)
strip_text(culture)
strip_text(cities)
strip_text(environment)
strip_text(law)
strip_text(science)
```

```python
def combine(data, date):
    our_long_string = ''
    index_arr = data.loc[data['date'] == date].index
    iter_list = ['description', 'title']
    for index in index_arr:
        for item in iter_list:     
            our_long_string += data[item][index] + '. '
    our_long_string.strip()
    
    return our_long_string
```

```python
# from https://pypi.org/project/rake-nltk/
# I decided to implement my own keyword extractor instead of keywords provided by the Guardian 
```

```python
from rake_nltk import Metric, Rake
def get_keywords(text):
    
    r = Rake(ranking_metric=Metric.DEGREE_TO_FREQUENCY_RATIO)
    r.extract_keywords_from_text(text)
    
    return r.get_ranked_phrases()
```

```python
#convert it to dictionary with values and its occurences
from collections import Counter
keys = get_keywords(c)
word_could_dict=Counter(keys)
wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(word_could_dict)

plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
#plt.show()
plt.savefig('yourfile.png', bbox_inches='tight')
plt.show()
```
